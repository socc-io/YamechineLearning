```python
#
# Pandas tutorial - 1 
#
# @author becxer
# @email becxer87@gmail.com
# @reference http://jsideas.net/python/2015/08/10/histogram_tutorial.html
#
from pandas import Series, DataFrame
import pandas as pd

print ("modules successfully imported")
```

```python
# 데이터를 불러옵니다, index는 자동으로 생성되고 가장 첫번째 row 가 csv  column 의 title입니다.
data = pd.read_csv('data/cvstore.csv')
data
```

```python
# 판다스의 기본 구조는 데이터 프레임 입니다.
# info 함수는 컬럼별 갯수와 타입, 메모리 크기 등을 보여줍니다.
data.info()
```

```python
# describe 는 숫자 컬럼의 갯수와 평균, 평균, 표준편차, 최소, 25,50,75% 값, 최대값 등을 보여줍니다.
data.describe()
```

```python
# hourly_wage 컬럼중에 6000 이상인 데이터를 불러옵니다.
data[data.hourly_wage > 6000]
```

```python
# Company 컬럼중에 CU가 포함된 데이터만 불러옵니다.
data[data.company.str.contains('CU')]
```

```python
# 데이터 중에 company 와 hourly_wage 컬럼만을 불러옵니다.
data_partial = data[['company', 'hourly_wage']]

# 위에 불러온 데이터 중에 4칸씩 row를 건너뛰면서 데이터 부분만 불러옵니다.
data_partial = data_partial[::4]
data_partial
```

```python
# 새로운 컬럼 good_money를 추가합니다. 
# 데이터 프레임은 map 함수를 포함하고 있으며, map함수에는 람다함수 를 대입가능합니다.
# hourly_wage 컬럼 데이터 중에 6000 이상인 데이터만 불러옵니다.

data_partial['good_money'] = data_partial.hourly_wage.map((lambda x : x > 6000))
data_partial
```

```python
#그래프를 그리기 위한 라이브러리를 불러옵니다.

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# 데이터 프레임에서 자동으로 pyplot을 불러들여서 그래프를 그릴수 있습니다.
# 막대 갯수가 40개인 그래프를 그립니다.
data_partial.hourly_wage.hist(bins=40)
```

```python
# hourly_wage 에 대해서 boxplot도 그릴수 있습니다.

data_partial.boxplot(column='hourly_wage')
```
